animation_group это группирование в каждом кадре    
то есть если мы ставим группировать по стране  
то у нас в каждом кадре будет группировка по стране  
в бар чаре это не заметно, так как тут группировка итак по континенту и все,  
а вот в скэттер плоте, если поставить группировать по континенту, то будет одна точка на графике,  
мы каждый фрейм будем группировать все данные в одну точку

```python
df = px.data.gapminder()

fig = px.bar(df, x="continent", y="pop", color="continent",
  animation_frame="year", animation_group="country", range_y=[0,4000000000], width=900, height=500)
fig.show()
```

```python
df = px.data.gapminder()
fig = px.scatter(df, x="gdpPercap", y="lifeExp", animation_frame="year", animation_group="country",
           size="pop", color="continent", hover_name="country", facet_col="continent",
           log_x=True, size_max=45, range_x=[100,100000], range_y=[25,90], width=900, height=500)
fig.show()
```

```python
df = px.data.gapminder().query("continent == 'Africa'")

fig = px.line(df, x="year", y="lifeExp", facet_col="country", facet_col_wrap=7,
              facet_row_spacing=0.04, # default is 0.07 when facet_col_wrap is used
              facet_col_spacing=0.04, # default is 0.03
              height=600, width=800,
              title="Life Expectancy in Africa")
fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
fig.update_yaxes(showticklabels=True)
fig.show()
```

Customizing lines in Plotly  
dash: set the dash style ("solid", "dot", "dash", "longdash", "dashdot", "longdashdot")  
shape: set how values are connected ("linear", "spline", "hv", "vh", "hvh","vhv")  

```python
# In this example, we’re updating a scatter plot named fig_ln
fig_ln.update_traces(patch={"line": {"dash": "dot",
                           "shape": "spline",
                           "width": 6}})
fig_ln.show()
```
